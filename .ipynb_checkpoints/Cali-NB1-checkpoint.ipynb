```python
import numpy as np
import pandas as pd
import io
import requests
from urllib.request import urlopen
from io import StringIO
import matplotlib.pyplot as plt
```

```python
## Read in raw data
#c = pd.read_csv("https://transparentcalifornia.com/export/school-districts/alameda/dublin-unified-2016.csv")
url = "https://transparentcalifornia.com/export/school-districts/alameda/dublin-unified-2016.csv"
url2 = "https://transparentcalifornia.com/export/school-districts/contra-costa/san-ramon-valley-unified-2016.csv"
url3 = "https://transparentcalifornia.com/export/school-districts/santa-clara/san-jose-unified-2016.csv"
url4 = "https://transparentcalifornia.com/export/school-districts/santa-clara/palo-alto-unified-2014.csv"
url5 = "https://transparentcalifornia.com/export/school-districts/san-francisco/san-francisco-unified-2016.csv"
url6 = "https://transparentcalifornia.com/export/school-districts/los-angeles/los-angeles-unified-2016.csv"
url7 = "https://transparentcalifornia.com/export/school-districts/alameda/oakland-unified-2016.csv"
url8 = "https://transparentcalifornia.com/export/school-districts/san-diego/san-diego-unified-2016.csv"

s = requests.get(url, verify=False).text
dublin = pd.read_csv(StringIO(s))

s2 = requests.get(url2, verify=False).text
srvusd = pd.read_csv(StringIO(s2))


s3 = requests.get(url3, verify=False).text
sjusd = pd.read_csv(StringIO(s3))

s4 = requests.get(url4, verify=False).text
pausd = pd.read_csv(StringIO(s4))

s5 = requests.get(url5, verify=False).text
sfusd = pd.read_csv(StringIO(s5))

s6 = requests.get(url6, verify=False).text
lausd = pd.read_csv(StringIO(s6))

s7 = requests.get(url7, verify=False).text
oakusd = pd.read_csv(StringIO(s7))

s8 = requests.get(url8, verify=False).text
sdusd = pd.read_csv(StringIO(s8))
```

```python
# Source http://dq.cde.ca.gov/dataquest/dqcensus/EnrGrdLevels.aspx?cds=19&agglevel=County&year=2016-17

dublin['Enrollment'] = 9965
srvusd['Enrollment'] = 32255
sjusd['Enrollment'] = 32454
pausd['Enrollment'] = 12527
sfusd['Enrollment'] = 60133
lausd['Enrollment'] = 633621
oakusd['Enrollment'] = 49760
sdusd['Enrollment'] = 128040
len(oakusd)
```

```python
total_employees = [len(dublin),len(srvusd),len(sjusd),len(pausd), 
                   len(sfusd), len(lausd), len(oakusd), len(sdusd)]
total_employees
e2e = {'enrollment' : [9965, 32255, 32454, 12527, 60133, 633621, 49760, 128040], 
       'employees'  : [len(dublin),len(srvusd),len(sjusd),len(pausd),
                      len(sfusd), len(lausd), len(oakusd), len(sdusd)],
       'schooldistrict' : ['Dbln', 'SRMN', 'SJose', 'Palo', 'SF', 'LA', 'OAK', 'SD']}
e2edf = pd.DataFrame(data=e2e)
e2edf.head(8)
```

```python

import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')

```

```python
e2edf.plot(x='schooldistrict', y='employees', kind='bar', color='green')
plt.show()
e2edf.plot(x='schooldistrict', y='enrollment', kind='bar', color='darkblue')
plt.show()
```

###### LA Unified school district is likely the largest school district in California. Is it too big to operate efficiently and meet the needs of students in LA? Has anyone looked into breaking it up into smaller and efficient school districts?

```python
e2edf['supsalary'] = [
                      sum(dublin[dublin['Job Title'] == "Superintendent"]['Total Pay']),
                      sum(srvusd[srvusd['Job Title'] == "Superintendent"]['Total Pay']),
                      sum(sjusd[sjusd['Job Title'] == "Superintendent"]['Total Pay']),
                      sum(pausd[pausd['Job Title'] == "SUPERINTENDENT"]['Total Pay']),
                      sum(sfusd[sfusd['Job Title'].str.contains("Superintendent")]['Total Pay']),
                      sum(lausd[lausd['Job Title'].str.contains("Superintendent")]['Total Pay']),
                      sum(oakusd[oakusd['Job Title'].str.contains("SUPERINTENDENT")]['Total Pay']),
                      sum(sdusd[sdusd['Job Title'].str.contains("Supt")]['Total Pay'])
                     ]



sum(dublin[dublin['Job Title'] == "Superintendent"]['Total Pay'])
sum(sjusd[sjusd['Job Title'] == "Superintendent"]['Total Pay'])
sum(srvusd[srvusd['Job Title'] == "Superintendent"]['Total Pay'])
sum(pausd[pausd['Job Title'] == "SUPERINTENDENT"]['Total Pay'])

avgsalary_dublin = sum(dublin['Total Pay'])/dublin['Enrollment'][0]
avgsalary_srvusd = sum(srvusd['Total Pay'])/srvusd['Enrollment'][0]
avgsalary_sjusd = sum(sjusd['Total Pay'])/sjusd['Enrollment'][0]
avgsalary_pausd = sum(pausd['Total Pay'])/pausd['Enrollment'][0]
avgsalary_sfusd = sum(sfusd['Total Pay'])/sfusd['Enrollment'][0]
avgsalary_lausd = sum(lausd['Total Pay'])/lausd['Enrollment'][0]
avgsalary_oakusd = sum(oakusd['Total Pay'])/oakusd['Enrollment'][0]
avgsalary_sdusd = sum(sdusd['Total Pay'])/sdusd['Enrollment'][0]
e2edf['payperstudent'] = [ avgsalary_dublin,  avgsalary_srvusd, avgsalary_sjusd, avgsalary_pausd,
                           avgsalary_sfusd, avgsalary_lausd, avgsalary_oakusd, avgsalary_sdusd]
```

###### Total salary is normalized by number of students enrolled in a school district. 
###### While SanJose Los Angeles has most students, 
###### Oakland has spent lowest average of staff salary per student. 
###### Does it mean Oakland is good at managing overall salary budget or is it a sign of funding shortage?

```python
matplotlib.style.use('ggplot')
e2edf.plot(x='schooldistrict', y='payperstudent', kind='bar', color='darkgreen', label="Tot Salary Per Student")

e2edf.plot(x='schooldistrict', y='supsalary', kind='bar', color='maroon', label="Sup. Salary")

plt.show()
```

###### Larger school districts tend to spent more on Suprintendent level positions. However there is no strong co-relation between enrolled students and salary paid to Suprintendents (Main or Assistant). The data strong suggests that higher pay suprintendent is not a sure shot for success. 

```python
e2edf['princisalary'] = [
                           sum(dublin[dublin['Job Title'].str.contains("Principal")]['Total Pay']),
                           sum(srvusd[srvusd['Job Title'].str.contains("Principal")]['Total Pay']),
                           sum(sjusd[sjusd['Job Title'].str.contains("Principal")]['Total Pay']),
                           sum(pausd[pausd['Job Title'].str.contains("PRINCIPAL")]['Total Pay']),
                           sum(sfusd[sfusd['Job Title'].str.contains("Principal")]['Total Pay']),
                           sum(lausd[lausd['Job Title'].str.contains("Principal")]['Total Pay']),
                           sum(oakusd[oakusd['Job Title'].str.contains("PRINCIPAL")]['Total Pay']),
                           sum(sdusd[sdusd['Job Title'].str.contains("Principal")]['Total Pay'])
                        ]
```

```python

e2edf.plot(x='schooldistrict', y='princisalary', kind='bar', color='darkblue', label="Principal. Salary")
plt.show()
```

```python
e2edf['teachersalary'] = [
                          sum(dublin[dublin['Job Title'].str.contains("Teacher")]['Total Pay'])/dublin['Enrollment'][0],
                          sum(srvusd[srvusd['Job Title'].str.contains("Teacher")]['Total Pay'])/srvusd['Enrollment'][0],
                          sum(sjusd[sjusd['Job Title'].str.contains("Teacher")]['Total Pay'])/sjusd['Enrollment'][0],
                          sum(pausd[pausd['Job Title'].str.contains("TEACHER")]['Total Pay'])/(pausd['Enrollment'][0]),
                          sum(sfusd[sfusd['Job Title'].str.contains("Teacher")]['Total Pay'])/sfusd['Enrollment'][0],
                          sum(lausd[lausd['Job Title'].str.contains("Teacher")]['Total Pay'])/lausd['Enrollment'][0],
                          sum(oakusd[oakusd['Job Title'].str.contains("TEACHER")]['Total Pay'])/oakusd['Enrollment'][0],
                          sum(sdusd[sdusd['Job Title'].str.contains("Teacher")]['Total Pay'])/(sdusd['Enrollment'][0])                 
                          ]

```

###### The teacher's salary is normalized by number of students enrolled in a school district. Teachers are most crucial part of an education system. It is interesting to note that Dublin and Palo Alto have similiar enrollment numbers, yet Palo Alto spends significantly higher on teacher salary per student than Dublin. 

```python
e2edf.plot(x='schooldistrict', y='teachersalary', kind='bar', color='maroon', label="Teacher's Salary")
plt.show()
```

###### This chart provides a fascinating insight about spending prioritization. Some school districts prioritize well paid teachers (Palo Alto), while other school districts don't do that (example Oakland)
